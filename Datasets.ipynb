# **Artificially Created Dataset for Linear Regression**

A dataset is a collection of data points used in machine learning. Each data point represents a specific instance or example and contains features or attributes that describe it.

Datasets are crucial for machine learning because they are the fuel that powers the learning process. Machine learning models learn patterns and relationships from the data they are trained on. The quality and quantity of data in a dataset directly impact the performance and accuracy of the resulting model.

An artificial dataset is a collection of data that is generated programmatically rather than collected from real-world events. These datasets are often created using algorithms to simulate the statistical properties of real data, making them useful for training and testing machine learning models when real data is scarce or sensitive.

https://en.wikipedia.org/wiki/Synthetic_data

## **Linear Regression with Synthetic Data**

This is an artificially created dataset for linear regression.

First, a seed is set for the random number generator to ensure reproducibility.

X is an array of m random numbers between 0 and 2.
y is created from a linear function of X with some added Gaussian noise. The formula is y = 3X + 4 + epsilon where epsilon is the Gaussian noise.

The dataset is designed to simulate a linear relationship between the independent variable X and the dependent variable y, but with some random noise to make it more realistic.

```python
import numpy as np
np.random.seed(42)  # to make this code example reproducible
m = 100  # m number of instances

# Create an array of the given shape and populate it with
# random samples from a uniform distribution
# over [0, 1). generates an array of shape (m, 1) filled with random values
# drawn from a uniform distribution over the interval ([0, 1))
X = 2 * np.random.rand(m, 1)  # column vector
```

```python
# The function that we use to generate the data is y = 4 + 3x1 + Gaussian noise
y = 3 * X + 4 + np.random.randn(m, 1)  # column vector

y[:5]
```

```python
correlation_matrix = np.corrcoef(X.T, y.T)
correlation = correlation_matrix[0, 1]

print(f"The correlation coefficient between X and y is: {correlation}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$X$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Slope:", lin_reg.coef_)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np # Added import statement

# Assuming X is your numpy array with independent variables
# and y is your numpy array with the dependent variable
df = pd.DataFrame(np.concatenate((X, y), axis=1)) # reshaped y to have two dimensions
sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt

y_predict = lin_reg.predict(X)

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.", label="Training Data" + " y = (3x + 4)")
plt.plot(X, y_predict, "r-", label="Linear Regression Model" + f" (y = {lin_reg.coef_[0][0]:.2f}x + {lin_reg.intercept_[0]:.2f})")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
plt.title("Linear Regression Predictions vs Actual Data")
plt.show()
```

```python
residuals = y - y_predict

# Plotting residuals
plt.figure(figsize=(6, 4))
plt.plot(residuals, "b.", label="Residuals")

plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel("$x_1$")
plt.ylabel("Residuals", rotation=0)
plt.grid()
plt.legend(loc="upper left")
plt.title("Residuals Plot")
plt.show()
```

```python
residuals.mean()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

## **Multiple Linear Regression Dataset (4 Features)**

## **Understanding the Dataset**

Multiple Independent Variables: The dataset has four independent variables (X[:, 0] to X[:, 3]), making it suitable for showcasing multiple linear regression.
Linear Relationship: The relationship between X and y is defined by a linear equation with added Gaussian noise.
Debugging Focus: Since the data is generated artificially, the "debugging" focuses on understanding the relationship between the variables and how well a model can capture it.

```python
import numpy as np
np.random.seed(42)
m = 100

# Independent variables
X = np.random.rand(m, 4)

# Calculates the linear combination of the columns in X first
y = 5 * X[:, 0] + -3 * X[:, 1] + 0.5 * X[:, 2] + - 0.8 * X[:, 3] + 5

# Then reshapes it to (100, 1), and finally adds the noise with the same shape
y = y.reshape(m, 1) + np.random.randn(m, 1)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

# You can add code for visualization and evaluation as needed
```

```python
y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
from sklearn.linear_model import SGDRegressor

lin_reg = SGDRegressor()
lin_reg.fit(X, y.ravel())

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive'],
    'max_iter': [500, 1000, 2000],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01]
}

# Create the GridSearchCV object
grid_search = GridSearchCV(SGDRegressor(), param_grid, cv=5)

# Fit the grid search to the data
grid_search.fit(X, y.ravel())

# Print the best parameters
print("Best parameters:", grid_search.best_params_)

# Use the best estimator
best_model = grid_search.best_estimator_
```

```python
# Create an SGDRegressor with adjusted hyperparameters

# best_model is equivalent to
# lin_reg = SGDRegressor(learning_rate='constant',
#                        max_iter=500,
#                        penalty='l2',
#                        alpha=0.001)


lin_reg = best_model

# Fit the model to the data
lin_reg.fit(X, y.ravel())

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)

# R-squared
r2 = r2_score(y, y_pred)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

## sklearn.datasets

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load the dataset
housing = fetch_california_housing()

# Display the DataFrame
# print(housing)

df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
df.head()
```

```python
df.info()
```

```python
def find_correlation(df):
  # compute standard correlation coefficient
  corr_matrix = df.corr()
  # fetch and return attribute correlates
  # with the median housing value
  return corr_matrix["MedHouseVal"].sort_values(
    ascending=False)

# find correlation coefficient
corr_coef = find_correlation(df)
print("Correlation Coefficient::", corr_coef)
```

```python
df.plot(kind='scatter',
             x="MedInc", y="MedHouseVal",
             alpha=0.1)
```

```python
df.hist(column='MedHouseVal',
             bins=50, figsize=(9,6))
```

```python
# Use pd.cut() method to create an income category attribute with 5 categories
df["IncCat"] = pd.cut(df["MedInc"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels=[1, 2, 3, 4, 5])

df["IncCat"].hist()
```

```python
# The prices have a slight tendency to go down when we go north.
# The ocean proximity attribute may be useful as well
# (the housing prices in the costal district are not too high).

import matplotlib.pyplot as plt
# plot
df.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
             s=df["Population"]/100, label="Population", figsize=(7,5),
             c="MedHouseVal", cmap=plt.get_cmap("jet"), colorbar=True,)
plt.legend()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

# Stratified sampling
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_index, test_index = next(split.split(df, df["IncCat"]))

# Get train and test datasets and removes the IncCat column from the train and test sets, as it was only needed for stratification.
train_set = df.loc[train_index].drop("IncCat", axis=1)
test_set = df.loc[test_index].drop("IncCat", axis=1)

# Create X and y for training
X_train = train_set.drop("MedHouseVal", axis=1)
y_train = train_set["MedHouseVal"].copy()

# Create X and y for validation
X_val = test_set.drop("MedHouseVal", axis=1)
y_val = test_set["MedHouseVal"].copy()

# Linear regression model for best fit
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predictions for training set
y_train_pred = lin_reg.predict(X_train)

# Mean Squared Error for training set
mse_train = mean_squared_error(y_train, y_train_pred)
print(f"Mean Squared Error (MSE) for training set: {mse_train:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared for training set
r2_train = r2_score(y_train, y_train_pred)
print(f"\nR-squared for training set: {r2_train:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")

# Predictions for validation set
y_val_pred = lin_reg.predict(X_val)

# Mean Squared Error for validation set
mse_val = mean_squared_error(y_val, y_val_pred)
print(f"\nMean Squared Error (MSE) for validation set: {mse_val:.2f}")

# R-squared for validation set
r2_val = r2_score(y_val, y_val_pred)
print(f"\nR-squared for validation set: {r2_val:.2f}")
```
