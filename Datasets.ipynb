# **Artificially Created Dataset for Linear Regression**

A dataset is a collection of data points used in machine learning. Each data point represents a specific instance or example and contains features or attributes that describe it.

Datasets are crucial for machine learning because they are the fuel that powers the learning process. Machine learning models learn patterns and relationships from the data they are trained on. The quality and quantity of data in a dataset directly impact the performance and accuracy of the resulting model.

## **Linear Regression with Synthetic Data**

This is an artificially created dataset for linear regression.

First, a seed is set for the random number generator to ensure reproducibility.

X is an array of m random numbers between 0 and 2.
y is created from a linear function of X with some added Gaussian noise. The formula is y = 3X + 4 + epsilon where epsilon is the Gaussian noise.

The dataset is designed to simulate a linear relationship between the independent variable X and the dependent variable y, but with some random noise to make it more realistic.

```python
import numpy as np
np.random.seed(42)  # to make this code example reproducible
m = 100  # m number of instances

# Create an array of the given shape and populate it with
# random samples from a uniform distribution
# over [0, 1). generates an array of shape (m, 1) filled with random values
# drawn from a uniform distribution over the interval ([0, 1))
X = 2 * np.random.rand(m, 1)  # column vector
```

```python
# The function that we use to generate the data is y = 4 + 3x1 + Gaussian noise
y = 3 * X + 4 + np.random.randn(m, 1)  # column vector

y[:5]
```

```python
correlation_matrix = np.corrcoef(X.T, y.T)
correlation = correlation_matrix[0, 1]

print(f"The correlation coefficient between X and y is: {correlation}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$X$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Slope:", lin_reg.coef_)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np # Added import statement

# Assuming X is your numpy array with independent variables
# and y is your numpy array with the dependent variable
df = pd.DataFrame(np.concatenate((X, y), axis=1)) # reshaped y to have two dimensions
sns.pairplot(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.", label="Training Data" + " y = (3x + 4)")
plt.plot(X, lin_reg.predict(X), "r-", label="Linear Regression Model" + f" (y = {lin_reg.coef_[0][0]:.2f}x + {lin_reg.intercept_[0]:.2f})")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
plt.title("Linear Regression Predictions vs Actual Data")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
from sklearn.linear_model import SGDRegressor

lin_reg = SGDRegressor()
lin_reg.fit(X, y.ravel())

lin_reg.coef_


import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.", label="Training Data" + " y = (3x + 4)")
plt.plot(X, lin_reg.predict(X), "r-", label="Linear Regression Model" + f" (y = {lin_reg.coef_[0]:.2f}x + {lin_reg.intercept_[0]:.2f})")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, 2, 0, 15])
plt.grid()
plt.legend(loc="upper left")
plt.title("SGDRegressor Predictions vs Actual Data")
plt.show()
```

## **Multiple Linear Regression Dataset (4 Features)**

Understanding the Dataset

Multiple Independent Variables: The dataset has four independent variables (X[:, 0] to X[:, 3]), making it suitable for showcasing multiple linear regression.
Linear Relationship: The relationship between X and y is defined by a linear equation with added Gaussian noise.
Debugging Focus: Since the data is generated artificially, the "debugging" focuses on understanding the relationship between the variables and how well a model can capture it.

```python
import numpy as np
np.random.seed(42)
m = 100

# Independent variables
X = np.random.rand(m, 4)

# Calculates the linear combination of the columns in X first
y = 5 * X[:, 0] + -3 * X[:, 1] + 0.5 * X[:, 2] + - 0.8 * X[:, 3] + 5

# Then reshapes it to (100, 1), and finally adds the noise with the same shape
y = y.reshape(m, 1) + np.random.randn(m, 1)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

# You can add code for visualization and evaluation as needed
```

```python
y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)

# R-squared
r2 = r2_score(y, y_pred)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
from sklearn.linear_model import SGDRegressor

lin_reg = SGDRegressor()
lin_reg.fit(X, y.ravel())

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)

# R-squared
r2 = r2_score(y, y_pred)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive'],
    'max_iter': [500, 1000, 2000],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01]
}

# Create the GridSearchCV object
grid_search = GridSearchCV(SGDRegressor(), param_grid, cv=5)

# Fit the grid search to the data
grid_search.fit(X, y.ravel())

# Print the best parameters
print("Best parameters:", grid_search.best_params_)

# Use the best estimator
best_model = grid_search.best_estimator_
```

```python
# Create an SGDRegressor with adjusted hyperparameters

# best_model is equivalent to
# lin_reg = SGDRegressor(learning_rate='constant',
#                        max_iter=500,
#                        penalty='l2',
#                        alpha=0.001)


lin_reg = best_model

# Fit the model to the data
lin_reg.fit(X, y.ravel())

print("Intercept:", lin_reg.intercept_)
print("Coefficients:", lin_reg.coef_)

y_pred = lin_reg.predict(X)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)

# R-squared
r2 = r2_score(y, y_pred)

# Mean Squared Error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
print("MSE measures the average squared difference between the predictions and the actual values.")

# R-squared
r2 = r2_score(y, y_pred)
print(f"\nR-squared: {r2:.2f}")
print("R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s).")
```
